1. *lr*, *betas* - parameters uses for **Adam** optimizer

3. *lookahead* - if **True** uses **Lookahead** wrapper on the optimizer, if **False** doesn't make any modifications
4. *lookahead_k* - *int* - internal parameter for **Lookahead** optimizer, specifies how many internal loop steps should be made
5. *eval_every* - specify how often you would log the results of the training (saves and outputs results every *eval_every* iterations)
6. *out_dir* - directory for results
7. *pretrained_clf_path* - path to the pretrained model *mnist.pth* required for the plot function and constructing of metrics

# Algorithms from the *extra_var_reduction* folder

### Import module *train_min_max* to run the experimetns:

```python
from extra_var_reduction import train_min_max
```

### Specify parameters for your experimetns
1. *lr* - learnign rate for the optimizer
2. *optim* - (*str*) could be equal to *sgd* or *adam* and specified which one to use
3. *betas* - parameter uses for **Adam** optimizer
4. *momentum* - parameter uses for **SGD** optimizer
5. *use_lookahead* - if **True** uses **Lookahead** wrapper on the optimizer, if **False** doesn't make any modifications
6. *batch_size* - size of the big batch batch_size
7. *mini_batch_size* - size of the smaller batch size, *batch_size* should be divisible by *mini_batch_size*
8. *eval_every* - specify how often you would log the results of the training (saves and outputs results every *eval_every* iterations)
9. *out_dir* - directory for results
10. *pretrained_clf_path* - path to the pretrained model *mnist.pth* required for the plot function and constructing of metrics

```python
seed = 13
modules_dir="./"
out_dir = modules_dir+"results/"
eval_every=100
epochs = 40
lr = 3e-4
batch_size = 1024
mini_batch_size = 256
grad_max_norm = 1
betas = (0.9, 0.999)
momentum = 0
optim = 'adam'
use_lookahead=True
```

### Run your experiments

```python
train_min_max.train(
    epochs=epochs,
    batch_size=batch_size,
    mini_batch_size=mini_batch_size,
    out_dir=out_dir,
    grad_max_norm=grad_max_norm,
    eval_every=eval_every,
    lr=lr,
    momentum=momentum,
    betas=betas,
    optim=optim,
    use_lookahead=use_lookahead,
    seed=seed,
    n_workers=2
)
```
